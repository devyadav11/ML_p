```python
# prompt: mark on each curve at mean photon number 1, 2, 3, 4, and 5 and also give the corresponding probability values of each curve  n = 1,2,3,4,5  (give us the values)

import math
import numpy as np
import matplotlib.pyplot as plt

def poisson_probability(n, mu):
    """Calculate the probability P(n) for a Poisson distribution."""
    return (math.exp(-mu) * mu**n) / math.factorial(n)

# Define mean photon numbers
mu_values = np.linspace(0, 10, 100)  # Smooth range for the x-axis

# Define photon numbers
n_values = [1, 2, 3, 4, 5]

# Create the plot
plt.figure(figsize=(10, 6))

# Plot for each photon number and mark points at mu = 1, 2, 3, 4, 5
for n in n_values:
    probabilities = [poisson_probability(n, mu) for mu in mu_values]
    plt.plot(mu_values, probabilities, label=f"n = {n}")
    for mu_mark in [1, 2, 3, 4, 5]:
        prob = poisson_probability(n, mu_mark)
        plt.plot(mu_mark, prob, marker='o', markersize=5, color='red')  # Mark points in red
        print(f"For n = {n}, mu = {mu_mark}, P(n) = {prob:.4f}")

# Customize the plot
plt.xlabel("Mean Photon Number (Î¼)")
plt.ylabel("Probability P(n)")
plt.title("Poisson Probability vs. Mean Photon Number for Different n")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

```python
import math

# Constants
h = 6.626e-34  # Planck's constant (Joule-seconds)
c = 3.0e8      # Speed of light (m/s)

def calculate_pulse_power(mu, wavelength, time_period):
    """
    Calculate pulse power (P) in watts given mean photon number (mu),
    wavelength (lambda), and time period (T).
    """
    return (mu * h * c) / (wavelength * time_period)

def convert_to_mW(power_watt):
    """Convert power from Watts to milliwatts (mW)."""
    return power_watt * 1e3  # 1W = 1000mW

def convert_to_dBm(power_mW):
    """Convert power from mW to dBm."""
    return 10 * math.log10(power_mW) if power_mW > 0 else float('-inf')  # Avoid log(0) error

# Example Parameters
mu = 5  # Mean photon number
wavelength = 1550e-9  # Wavelength in meters (e.g., 1550 nm for telecom fiber)
time_period = 3e-9  # Time period in seconds (1 ns pulse duration)

# Compute power in Watts
power_watt = calculate_pulse_power(mu, wavelength, time_period)

# Convert to mW and dBm
power_mW = convert_to_mW(power_watt)
power_dBm = convert_to_dBm(power_mW)

# Display results
print(f"Pulse Power: {power_watt:.6e} W")
print(f"Pulse Power: {power_mW:.10f} mW")
print(f"Pulse Power: {power_dBm:.2f} dBm")
```
